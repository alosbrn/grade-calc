```python
import pandas as pd
from time import sleep
```

```python
# Create variables for grade input/output.
grades_in = "data/Grades_Short.csv"
grades_out = "data/Grades_Short_Letter.csv"
```

```python
# Read grades from provided .csv file.
grades = pd.read_csv(grades_in)
```

```python
# Filter grades to only include columns containing numeric values. 
# Assumes student ID is found in the first column.
# Allows the code following to average a variable number of grades.
grade_filter = grades.select_dtypes(include="number").columns[1:]
```

```python
# Calculate the average of the provided grades. 
# Adds the final grade (rounded to the first decimal) to our grade sheet as a new column.
grades["Final_Grade"] = grades[grade_filter].mean(axis=1).round(1)
```

```python
# Define a function for converting a final grade in numeric form to a letter grade.
def get_letter_grade(final_grade):
    if final_grade >= 90:
        return "A+"
    elif final_grade >= 80:
        return "A"
    elif final_grade >= 70:
        return "B"
    elif final_grade >= 60:
        return "C"
    elif final_grade > 55:
        return "D"
    else:
        return "F"
```

```python
# Create a new column containing the calculated letter grade.
grades["Letter_Grade"] = grades["Final_Grade"].apply(get_letter_grade)
```

```python
# Print an advancing save message, with pauses to simulate the program "working". 
# This is for demo purposes only. 
t = 0
while t != 3:
    save_msg = "Saving." + ("." * t)
    print(f"{save_msg}")
    sleep(1)
    t += 1
```

```python
# Save the modified grades to a new .csv file, excludes index as to match input file.
grades.to_csv(grades_out, index=False)
print(f"Saved as '{grades_out}'!")
```
